# <center>20. Valid Parentheses<center>

<div>

Given a string `s` containing just the characters `'('`, `')'`, `'{'`, `'}'`, `'['` and `']'`, determine if the input string is valid.

An input string is valid if:

1.  Open brackets must be closed by the same type of brackets.
2.  Open brackets must be closed in the correct order.

**Example 1:**

<pre>**Input:** s = "()"
**Output:** true
</pre>

**Example 2:**

<pre>**Input:** s = "()[]{}"
**Output:** true
</pre>

**Example 3:**

<pre>**Input:** s = "(]"
**Output:** false
</pre>

**Example 4:**

<pre>**Input:** s = "([)]"
**Output:** false
</pre>

**Example 5:**

<pre>**Input:** s = "{[]}"
**Output:** true
</pre>

**Constraints:**

*   `1 <= s.length <= 10<sup>4</sup>`
*   `s` consists of parentheses only `'()[]{}'`.

</div>

```python
%%java

// 1ms, O(1)
class Solution {
    public boolean isValid(String s) {
        Stack<Character> stack = new Stack<>();
        for(int i=0;i<s.length();i++){
            switch (s.charAt(i)){
                case '(':
                    stack.push(s.charAt(i));
                    break;
                case '[':
                    stack.push(s.charAt(i));
                    break;
                case '{':
                    stack.push(s.charAt(i));
                    break;
                case ')':
                    if(!stack.isEmpty()&&stack.peek()=='(')
                        stack.pop();
                    else
                        return false;
                    break;
                case ']':
                    if(!stack.isEmpty()&&stack.peek()=='[')
                        stack.pop();
                    else
                        return false;
                    break;
                case '}':
                    if(!stack.isEmpty()&&stack.peek()=='{')
                        stack.pop();
                    else
                        return false;
                    break;
            }
        }
        if(stack.size() != 0) return false;
        return true;
    }
}
```

```python
%%python
# simple way, same concept, but not too many if
class Solution(object):
    def isValid(self, s):
        # initialize our stack and hashtable
        stack = []
        hashTable = {
            "]":"[",
            ")":"(",
            "}":"{"
        }
        # Loop through all of the parentheses
        for par in s:
            # If the current parentheses is in the "values" section of our hashtable, add it to the stack
            if par in hashTable.values():
                stack.append(par)
            # If the current parentheses is in the "keys" section of our hashtable, pop the stack and compare the popped value to the value given the key
            elif par in hashTable.keys():
                # Note that we also check if the stack is empty. If the stack is currently empty, then we got a close parentheses with no open. Return false!
                if stack == [] or stack.pop() != hashTable[par]:
                    return False
            # Note that if we get a character that isn't in either, then the input isn't valid and we should just return False
            else:
                return False
        # If we go the entire string and have nothing in the stack, then we return true. If we still have chars in the stack, then return false!
        return stack == []
```
