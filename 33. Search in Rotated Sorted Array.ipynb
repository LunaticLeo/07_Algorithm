# <center>33. Search in Rotated Sorted Array</center>

There is an integer array `nums` sorted in ascending order (with **distinct** values).

Prior to being passed to your function, `nums` is **possibly rotated** at an unknown pivot index `k` (`1 <= k < nums.length`) such that the resulting array is `[nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]]` (**0-indexed**). For example, `[0,1,2,4,5,6,7]` might be rotated at pivot index `3` and become `[4,5,6,7,0,1,2]`.

Given the array `nums` **after** the possible rotation and an integer `target`, return _the index of_ `target` _if it is in_ `nums`_, or_ `-1` _if it is not in_ `nums`.

You must write an algorithm with `O(log n)` runtime complexity.

**Example 1:**

**Input:** nums = \[4,5,6,7,0,1,2\], target = 0
**Output:** 4

**Example 2:**

**Input:** nums = \[4,5,6,7,0,1,2\], target = 3
**Output:** -1

**Example 3:**

**Input:** nums = \[1\], target = 0
**Output:** -1

**Constraints:**

*   `1 <= nums.length <= 5000`
*   `-104 <= nums[i] <= 104`
*   All values of `nums` are **unique**.
*   `nums` is an ascending array that is possibly rotated.
*   `-104 <= target <= 104`

## Solution: Loop to find biggest and then binary search

Time complexity:  O(n+ log(n))

Space complexity: O(1)

```python
%java
// time: 0ms, 100%
// memory: 41.43MB, 97.57%

class Solution {
    public int search(int[] nums, int target) {
        if(nums.length ==0){
            if(nums[0] == target){
                return 0;
            }else{
                return -1;
            }
        }

        int n = nums.length;
        int pivot = -1; // assume and find the index of biggest number in nums as pivot

        for (int i = 0; i < n; i++) {
            if (nums[i] >= nums[(i - 1 + n) % n] && nums[i] >= nums[(i + 1) % n]) {
                pivot = i;
                break;
            }
        }

        int a, b;

        if (target > nums[pivot]) {
            return -1;
        } else if (target > nums[n - 1]) {
            a = 0;
            b = pivot;
        } else {
            a = (pivot + 1) % n;
            b = n - 1;
        }

        while (a < b) {
            if (nums[(a + b) / 2] == target) {
                return (a + b) / 2;
            } else if (nums[(a + b) / 2] < target) {
                a = (a + b) / 2 + 1;
            } else {
                b = (a + b) / 2 - 1;
            }
        }

        if (nums[a] == target) {
            return a;
        } else {
            return -1;
        }

    }
}
```

This method use binary search to find pivot

Time complexity: O(2* log(n))

Space complexity: O(1)

```python
%java
// time: 0ms, 100%
// memory: 41.43MB, 97.57%

class Solution {
    public int search(int[] nums, int target) {
        int n = nums.length;

        int left = 0, right = n - 1;

        while (left <= right) {
            int mid = (left + right) / 2;

            if (nums[mid] > nums[n - 1]) {
                left = mid + 1;
            } else {
                right = mid - 1;
            }
        }

        int pivot = (left - 1 + n) % n; // pivot is the index of biggest number in nums as pivot

        if (target > nums[pivot]) {
            return -1;
        } else if (target > nums[n - 1]) {
            left = 0;
            right = pivot;
        } else {
            left = (pivot + 1) % n;
            right = n - 1;
        }

        while (left <= right) {
            if (nums[(left + right) / 2] == target) {
                return (left + right) / 2;
            } else if (nums[(left + right) / 2] < target) {
                left = (left + right) / 2 + 1;
            } else {
                right = (left + right) / 2 - 1;
            }
        }
        return -1;
    }
}
```

## Solution: one binary search

The method only use binary search once, it clearify the case of each subarray's location and changes the choose method of mid

Time complexity:  O(log(n))

Space complexity: O(1)

```python
%java
// time: 0ms, 100%
// memory: 41.43MB, 97.57%

class Solution {
    public int search(int[] nums, int target) {
        int n = nums.length;
        int left = 0, right = n - 1;
        
        while (left <= right) {
            int mid = left + (right - left) / 2;
            
            // Case 1: find target
            if (nums[mid] == target) {
                return mid;
            }
            
            // Case 2: subarray on mid's left is sorted
            else if (nums[mid] >= nums[left]) {
                if (target >= nums[left] && target < nums[mid]) {
                    right = mid - 1;
                } else {
                    left = mid + 1;
                }
            }
            
            // Case 3: subarray on mid's right is sorted
            else {
                if (target <= nums[right] && target > nums[mid]) {
                    left = mid + 1;
                } else {
                    right = mid - 1;
                }
            }
        }
        
        return -1;
    }
}
```
